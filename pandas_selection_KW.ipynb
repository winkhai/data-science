# using  .loc (label/name based)

```python
tips_data.loc[20]
```

```python
tips_data.loc[[20]]
```

```python
tips_data.loc[[5, 25]]
```

```python
tips_data.loc[30:35]
```

```python
tips_data.loc[:, 'total_bill'].head()
```

```python
tips_data.loc[:, ['total_bill']].head()
```

```python
tips_data.loc[:, ['total_bill', 'day']]
```

```python
df = tips_data.set_index('sex', inplace=False)
df.head()
```

```python
df.loc['Female']
```

```python
df.loc['Male']
```

# using .iloc (integer/position based) 

```python
tips_data.head()
```

```python
tips_data.iloc[1]
```

```python
tips_data.iloc[[1]]
```

```python
tips_data.iloc[1:3]
```

```python
tips_data.iloc[:, 4]
```

```python
tips_data.iloc[:, [4]].head()
```

```python
tips_data.iloc[0:2, [3]].head()
```

```python
tips_data.iloc[0:2, [2, 1]].head()
```
